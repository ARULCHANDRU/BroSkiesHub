```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('sales_data.csv')

# Display the first 5 rows of the DataFrame
print("First 5 rows of the data:")
print(df.head())

# Get a summary of the DataFrame, including data types and non-null values
print("\nDataFrame Info:")
df.info()
```

```python
# Create a new column 'Total_Sales' by multiplying Price and Quantity
df['Total_Sales'] = df['Price'] * df['Quantity']

print("\nData with Total_Sales column:")
print(df.head())

# Use groupby() to find the sum of sales for each product
sales_by_product = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)

print("\nTotal sales by product:")
print(sales_by_product)
```

```python
# Create a bar chart of total sales by product
print("\nGenerating bar chart...")
plt.figure(figsize=(10, 6)) # Adjusts the size of the plot
sales_by_product.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'gold', 'orchid'])
plt.title('Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45) # Rotates the x-axis labels to prevent overlap
plt.tight_layout() # Adjusts plot to ensure everything fits without overlapping
plt.show()
```
